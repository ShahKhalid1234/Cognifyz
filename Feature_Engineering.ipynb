```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('Dataset .csv')

# 1. Extract additional features
data['Restaurant Name'] = data['Restaurant Name'].apply(len)
data['Address_Length'] = data['Address'].apply(len)

# 2. Create new features by encoding categorical variables
data['Has_Table_Booking'] = data['Has Table booking'].map({'Yes': 1, 'No': 0})
data['Has_Online_Delivery'] = data['Has Online delivery'].map({'Yes': 1, 'No': 0})

# Plotting the features
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 8))

# Plot 1: Restaurant Name Length
axes[0, 0].hist(data['Restaurant Name'], bins=20, color='skyblue')
axes[0, 0].set_title('Restaurant Name Length')

# Plot 2: Address Length
axes[0, 1].hist(data['Address_Length'], bins=20, color='salmon')
axes[0, 1].set_title('Address Length')

# Plot 3: Has Table Booking
axes[1, 0].bar(data['Has_Table_Booking'].value_counts().index, data['Has_Table_Booking'].value_counts().values, color=['lightgreen', 'lightcoral'])
axes[1, 0].set_xticks([0, 1])
axes[1, 0].set_xticklabels(['No', 'Yes'])
axes[1, 0].set_title('Has Table Booking')

# Plot 4: Has Online Delivery
axes[1, 1].bar(data['Has_Online_Delivery'].value_counts().index, data['Has_Online_Delivery'].value_counts().values, color=['lightgreen', 'lightcoral'])
axes[1, 1].set_xticks([0, 1])
axes[1, 1].set_xticklabels(['No', 'Yes'])
axes[1, 1].set_title('Has Online Delivery')

plt.tight_layout()
plt.show()
```
